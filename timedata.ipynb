# Time data  Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
radiodata=pd.read_csv('DataRadio.csv')

```

```python
plt.figure(figsize =(7,5))
ax = ax=sns.countplot(x='n4modepaym',data=radiodata,hue='n3gender')
plt.xticks(size =12)
plt.xlabel('Payment', size = 14)
plt.yticks(size = 12)
plt.ylabel('Numbers in payment method', size = 12)
plt.title("payment method percentages and count", size = 16)

payment = ['Mpesa', 'Cash', 'Credit', 'Nhif'] 
pay=[]

plt.legend(title='Gender', loc='upper right', labels=['Male', 'Female'])
ax.set_yticklabels(pay)
ax.set_xticklabels(payment)

total = len(radiodata)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.show()
```

```python
radiodata['date_time_in']=radiodata['date']+' '+radiodata['btime']
radiodata['date_time_out']=radiodata['date1']+ ' '+radiodata['btimeout']
```

```python
radiodata=radiodata.drop(columns=['date','btime','btimeout','date1','daysdiff','diffhrs'],axis=1)
```

```python
radiodata['date_time_out']=pd.to_datetime(radiodata['date_time_out'],errors='coerce')
radiodata['date_time_in']=pd.to_datetime(radiodata['date_time_in'],errors='coerce')
```

```python
radiodata['date_time_difference']=radiodata['date_time_out']-radiodata['date_time_in']
```
